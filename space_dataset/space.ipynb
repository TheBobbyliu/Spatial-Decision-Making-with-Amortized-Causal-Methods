#### cdcsvi_nohsdp_poverty_cont

Collection: Social Vulnerability & Welfare

Spatial unit / region: Census tract, Texas (USA)

Treatment → Outcome: nohsdp (share of adults with no high-school diploma) → poverty (poverty rate)

Treatment type: Continuous (real-valued exposure)

Size/shape: 6,828 tracts, 21,585 edges, 11 observed covariates

Spatial info: “Normal coordinates” (lat/lon) and adjacency graph

What you get in a SpaceDataset: training data, counterfactualsmoothness & confounding scores. 
arXiv

#### cdcsvi_nohsdp_poverty_disc

Same collection/unit/region as above (TX tracts).

Treatment → Outcome: same variables (nohsdp → poverty)

Treatment type: Binary (discretized treatment)

Size/shape: also listed with 11 covariates (same environment, different treatment coding). 
arXiv

#### zcta_age_broadband_cont

Collection: Socioeconomic Status & Broadband Usage

Spatial unit / region: ZCTA (ZIP Code Tabulation Area), USA

Treatment → Outcome: age (age-composition measure from Census) → broadband (broadband usage/adoption at ZCTA level, built from Form-477/ACS sources used by SpaCE)

Treatment type: Continuous

Size/shape: 30,383 ZCTAs, 174,345 edges, 16 observed covariates

Spatial info: coordinates + adjacency graph. 
arXiv

#### zcta_income_broadband_disc

Collection: Socioeconomic Status & Broadband Usage

Spatial unit / region: ZCTA, USA

Treatment → Outcome: income (income measure from Census) → broadband (ZCTA broadband usage/adoption)

Treatment type: Discrete (binned; SpaCE lists it as “discrete”)

Size/shape: same ZCTA environment: 30,383 nodes, 174,345 edges, 16 covariates (treatment recoded)

Spatial info: coordinates + adjacency graph.

```python
import numpy as np
import rasterio
import matplotlib.pyplot as plt
def preprocess(img):
    x = img.copy()
    # percentile stretch ignoring NaNs
    lo, hi = np.nanpercentile(x, [2, 98])
    x = np.clip((x - lo) / (hi - lo + 1e-6), 0, 1)
    x *= 255
    x = np.int64(x)
    x = np.moveaxis(x, 0, -1)
    return x

taskname = 'cdcsvi_nohsdp_poverty_disc'
src = rasterio.open("downloads/{}/{}.tif".format(taskname, 0)).read()
dst = preprocess(src)
plt.imshow(dst)
```
